```python
import cv2
import os
```

Load all of the images names

```python
images_name = os.listdir('data/jpg')
len(images_name)
```

Resize shape of all the images into 100*100 pixels

```python
for img_name in images_name:
    img = cv2.imread('data/jpg/'+img_name)
    img = cv2.resize(img, (100, 100))
    cv2.imwrite('data/compressed/'+img_name, img)
```

Load .mat file for separate index of all the train, test and valid images

```python
from scipy.io import loadmat
mat = loadmat('data/setid.mat')
mat
```

```python
train_id = mat['tstid'].reshape(-1)
valid_id = mat['valid'].reshape(-1)
test_id  = mat['trnid'].reshape(-1)
```

Find images classes per id

```python
labels = loadmat('data/imagelabels.mat')
labels 
```

This means that the the image with id 1 is class 77 and the image with id 8189 belongs to to class 62

```python
labels = labels['labels']
labels = labels.reshape(-1)
labels.shape
```

Read and write method to convert all images to binary file instead of png

```python
import numpy as np

def write_data(file_path, X, Y):
    with open(file_path, 'wb') as f:

        header = np.array([len(X), X.shape[1], X.shape[2], X.shape[3]], dtype=np.uint32)
        header.tofile(f)
    
        X.tofile(f)
        
        Y.tofile(f)


def read_data(file_path):
    with open(file_path, 'rb') as f:
        
        header = np.fromfile(f, dtype=np.uint32, count=4)
        num_samples, rows, columns, num_channels = header
    
        X = np.fromfile(f, dtype=np.uint8, count=num_samples * rows * columns * num_channels)
        X = X.reshape(num_samples, rows, columns, num_channels)
    
        Y = np.fromfile(f, dtype=np.uint8)
        return X, Y
```

Save the train part od dataset by its lables

```python
def set_len(value):
    if value < 10:
        return '0000'
    if value < 100:
        return '000'
    if value < 1000:
        return '00'
    if value < 10000:
        return '0'

x_train = np.ndarray((len(train_id), 100, 100, 3)).astype(np.uint8)
y_train = np.ndarray(len(train_id)).astype(np.uint8)

i = 0
for index in train_id:
    image = cv2.imread('data/compressed/' + 'image_' + set_len(index) + str(index) + '.jpg')
    y_train[i] = labels[index - 1]
    x_train[i] = image
    i += 1
```

Make sure that data saved without any problem when we have to read

```python
import matplotlib.pyplot as plt
print(y_train[6148])
plt.imshow(x_train[6148])
```

```python
write_data(file_path='data/oxfords_flowers_train.bin', X=x_train, Y=y_train)
```

```python
X, Y = read_data(file_path='data/oxfords_flowers_train.bin')
print(Y[6148])
plt.imshow(X[6148])
```

Results are the same, so the binary file is alright, Let's do the same thing with the test and valid sets

```python
x_test = np.ndarray((len(test_id), 100, 100, 3)).astype(np.uint8)
y_test = np.ndarray(len(test_id)).astype(np.uint8)

i = 0
for index in test_id:
    image = cv2.imread('data/compressed/' + 'image_' + set_len(index) + str(index) + '.jpg')
    y_test[i] = labels[index - 1]
    x_test[i] = image
    i += 1

write_data(file_path='data/oxfords_flowers_test.bin', X=x_test, Y=y_test)
```

```python
x_valid = np.ndarray((len(valid_id), 100, 100, 3)).astype(np.uint8)
y_valid = np.ndarray(len(valid_id)).astype(np.uint8)

i = 0
for index in valid_id:
    image = cv2.imread('data/compressed/' + 'image_' + set_len(index) + str(index) + '.jpg')
    y_valid[i] = labels[index - 1]
    x_valid[i] = image
    i += 1

write_data(file_path='data/oxfords_flowers_valid.bin', X=x_valid, Y=y_valid)
```
